```python
interpolado = grade_celular

for coluna in setores.iloc[:,1:-1]:
    ct_area_interp_gdf = area_interpolate(source_df=setores, 
                                          target_df=grade_celular,
                                          extensive_variables=[coluna])
    
    interpolado[coluna] = ct_area_interp_gdf[coluna]
```

```python
interpolado['avg_inv_hou'] = interpolado['DR_V002']/(interpolado['B_V001']+0.01)
```

```python
interpolado['dr_15'] = (interpolado[['P13_V022', 'P13_V036', 'P13_V037', 'P13_V038', 'P13_V039', 'P13_V040',
       'P13_V041', 'P13_V042', 'P13_V043', 'P13_V044', 'P13_V045', 'P13_V046',
       'P13_V047', 'P13_V048']].sum(axis=1))/(interpolado['P13_V001']+0.01)
```

```python
interpolado['dr_65'] = (interpolado[['P13_V099', 'P13_V100', 'P13_V101', 'P13_V102',
       'P13_V103', 'P13_V104', 'P13_V105', 'P13_V106', 'P13_V107', 'P13_V108',
       'P13_V109', 'P13_V110', 'P13_V111', 'P13_V112', 'P13_V113', 'P13_V114',
       'P13_V115', 'P13_V116', 'P13_V117', 'P13_V118', 'P13_V119', 'P13_V120',
       'P13_V121', 'P13_V122', 'P13_V123', 'P13_V124', 'P13_V125', 'P13_V126',
       'P13_V127', 'P13_V128', 'P13_V129', 'P13_V130', 'P13_V131', 'P13_V132',
       'P13_V133', 'P13_V134']].sum(axis=1))/(interpolado['P13_V001']+0.01)
```

```python
interpolado['literacy'] = interpolado['P01_V078']/(interpolado['B_V001']+0.01)
```

```python
interpolado['inad_sewag'] = (interpolado[['D01_V019', 'D01_V020', 'D01_V021', 'D01_V022', 'D01_V023']].sum(axis=1))/(interpolado['B_V001']+0.01)
```

```python
interpolado['inad_water'] = (interpolado['D01_V015'])/(interpolado['B_V001']+0.01)
```

```python
interpolado['inap_waste'] = (interpolado[['D01_V039','D01_V040', 'D01_V041', 'D01_V042']].sum(axis=1))/(interpolado['B_V001']+0.01)
```

```python
indicators = interpolado[['ID','geometry','avg_inv_hou', 'p_per_hous', 'dr_15', 'dr_65', 'sex_ratio', 'literacy',
       'inad_sewag', 'inad_water', 'inap_waste']]
```

```python
indicators
```

```python
scaler = StandardScaler()
indicators.iloc[:,2:] = scaler.fit_transform(indicators.iloc[:,2:])
indicators
```

```python
indicators.iloc[:,2:]
```

```python
corr_df = indicators.corr()

corr_df.style.background_gradient(cmap='Spectral')
```

```python
dados = indicators.iloc[:,2:].to_numpy()
dados
```

```python
fig, axes = plt.subplots(1, 1, figsize=(8,5))

fig.set_facecolor('white')
dendrograma = dendrogram(linkage(dados, method='ward'), ax=axes, p = 3, truncate_mode = 'level', color_threshold = 0.9)
plt.xlabel('Número de feições em cada cluster', fontsize = 14)
plt.ylabel('Distância euclidiana', fontsize = 14);
plt.title('Dimensão Ambiental', fontsize = 14,weight='bold')
plt.show()
```

```python
ag_model = AgglomerativeClustering(n_clusters = 6, linkage='ward', affinity='euclidean')
ag_model.fit(dados)
```

```python
unique, counts = np.unique(ag_model.labels_, return_counts=True)

result = np.column_stack((unique, counts)) 
print (result)
```

```python
indicators.plot(column= indicators['Clusters_6'], legend=True, categorical=True)
```
